```python
from tkinter import *
import math
import tkinter.font as font
root = Tk()
root.title('Calculator')
root.geometry('300x389')

screen= Entry(root, font=('Helvetica',17), justify='right')
screen.grid(row=0,column=0,columnspan=5, padx=5, ipadx=10,ipady=10)
screen.insert(0,'0')
keyboard =  {'MC': 0, 'MR': 0, 'MS': 0, 'M+': 0, 'M-': 0, '<-': 0, 'CE': 0, 'C': 4, '+/-': 6, 'sqrt': 5, '7': 1, 
             '8': 1,'9': 1, '/': 2, '%': 5, '4': 1, '5': 1, '6': 1,
             '*': 2,'1/x': 5, '1': 1, '2': 1, '3': 1, '-': 2, '=': 3, '0': 1, '': 0, '.': 7, '+': 2}
text_display=""
num1=0
num2=0
action=''
dot=False
def number_press(event):
    global text_display
    pressed=event.widget['text']
    if text_display=="" or text_display=='0':
        if pressed=='0':
            text_display = '0'
        else:
            if text_display=='0':
                text_display=""
            text_display += pressed
    else: 
        if text_display=='0':
            text_display=""
        text_display += pressed
    screen.delete(0,'end')
    screen.insert(0,text_display)
def action_press(event):
    global num1
    num1=float(screen.get())
    global action
    action=event.widget['text']
    global text_display
    text_display=""
    
def finish_press(event):
    global num2
    num2= float(screen.get())
    if action=='+':
        result=num1+num2
    elif action=='-':
        result=num1-num2
    elif action=='*':
        result=num1*num2
    else:
        if num2==0:
            result='NaN'
        else:
            result=num1/num2
    screen.delete(0,'end')
    screen.insert(0,result)
    
def clear_press(event):
    global text_display
    text_display=""
    global num1
    num1=0
    global num2
    num2=0
    global action
    action=''
    screen.delete(0,'end')
    screen.insert(0,'0')
    
def monoaction_press(event):
    global num1
    num1=float(screen.get())
    global action
    action=event.widget['text']
    if action=='1/x':
        if num1!=0:
            result=1/num1
        else:
            result='NaN'
    elif action=='sqrt':
        if num1<0:
            result='NaN'
        else:
            result=math.sqrt(num1)
    elif action=='%':
        result=num1*100
    global text_display
    text_display=''
    screen.delete(0,'end')
    screen.insert(0,result)
def reverse_press(event):
    global num1
    num1=float(screen.get())
    result=-num1
    screen.delete(0,'end')
    screen.insert(0,result)
def float_press(event):
    global text_display
    global dot
    if not dot:
        text_display+="."
    dot=True
#gồm có 30 phím
keys=list(keyboard.keys())
row=1
col=0
for i in range (len(keys)):
    if i%5==0 and i!=0:
        row+=1
    if i==26:
        continue
    if i==25:
        btn=Button(root, text=keys[i], font=('Helvetica',10,'bold'), height=2, width=13)
        btn.grid(row=row, column=i%5, columnspan=2,padx=5,pady=5)
    elif i==24:
        btn=Button(root, text=keys[i], font=('Helvetica',10,'bold'), height=5, width=5)
        btn.grid(row=row, column=i%5, rowspan=2,padx=5,pady=5, ipady=3)
    else:
        btn=Button(root, text=keys[i], font=('Helvetica',10, 'bold'), height=2, width=5)
        btn.grid(row=row, column=i%5, padx=5,pady=5)  
    if keyboard[keys[i]]==1:
        btn.bind("<Button-1>", number_press)
    elif keyboard[keys[i]]==2:
        btn.bind("<Button-1>", action_press)
    elif keyboard[keys[i]]==3:
        btn.bind("<Button-1>", finish_press) 
    elif keyboard[keys[i]]==4:
        btn.bind("<Button-1>", clear_press)  
    elif keyboard[keys[i]]==5:
        btn.bind("<Button-1>", monoaction_press)
    elif keyboard[keys[i]]==6:
        btn.bind("<Button-1>", reverse_press)
    elif keyboard[keys[i]]==7:
        btn.bind("<Button-1>", float_press)
root.mainloop()    
```
