```python
kl=dt.pivot_table(values='GDP (millions of US$)',index='Continent',aggfunc='sum')
display(kl)
print('='*50)
al=dt.pivot_table(values='GDP (millions of US$)',index='Continent',aggfunc='mean')
display(al)
```

```python
dm=pd.concat([kl,al],axis=1)
dm.columns=['Tổng GDP','TBC GDP']
dm
```
