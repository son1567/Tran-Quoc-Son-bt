```python
data['total']=data['Quantity']*data['UnitPrice']
#So luong don hang
No_invoices=data['InvoiceNo'].unique().size
#Tong gia tri tat ca don hang
total=data['total'].sum()
print("So luong hoa don ban ra:"+str(data['InvoiceNo'].unique().size))
print("Tong doanh thu:"+str(total))
```
