```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
print('thanh cong')
```

```python
dt=pd.read_csv('subset-covid-data.csv')
dt
```

```python
dt.info()
```

```python
# du lieu duoc thong ke cho nhung ngay nao
dt.date.value_counts()
```

```python
#loc du lieu nhieu
cleaned_dt=dt[dt.date=='2020-04-12']
```

```python
#ve bieu do phan bo so luong cac ca nhiem o cac quoc gia
print('Trung binh so ca mac mơi: '+str(cleaned_dt.cases.mean()))
print('Trung vi cua nhung ca mac moi: '+str(cleaned_dt.cases.median()))
plt.hist(cleaned_dt.cases,bins=200)
plt.title('Phan bo so ca mac')
plt.xlabel('So ca mac moi')
plt.ylabel('So luong cac quoc gia')
```

```python
print('Tong so ca nhiem va so ca cua cac chau luc')
cleaned_dt.groupby('continent')['cases','deaths'].sum()
```

```python
# quoc gia co so ca nhiem moi cao nhat
dt=dt.sort_values('cases',ascending=False)
dt.head(5)
```

```python
#top 5 quoc gia co so ca tu vong 
dt=dt.sort_values('deaths',ascending=False)
dt.head(5)
```
