```python
dt.loc[:,'type_of_land'].unique()
```

```python
dt.loc[dt.loc[:,'type_of_land']=='Bán nhà riêng\n ','type_of_land']='Bán nhà riêng'
dt.loc[dt.loc[:,'type_of_land']=='Bán nhà mặt phố\n ','type_of_land']='Bán nhà mặt phố'
dt.loc[dt.loc[:,'type_of_land']=='Bất động sản khác\n ','type_of_land']='Bất động sản khác'
dt.loc[:,'type_of_land'].unique()
```

```python
dt[(dt.loc[:,'type_of_land']=='Bán nhà riêng') & (dt.loc[:,'ward_name']=='Phường Trung Liệt') | (dt.loc[:,'ward_name']=='Phường Khâm Thiên')]
```

```python
dt[(dt.loc[:,'bedroom']>=3) & (dt.loc[:,'land_certificate']=='Sổ đỏ')].loc[:,['address','price','house_direction','balcony_direction','bedroom','land_certificate']]
```

```python
print(' Giá trung bình cộng của mỗi loại nhà đất')
display(dt.groupby('type_of_land')['price'].mean())
print('='*30)
print(' Giá nhỏ nhất của mỗi loại nhà đất')
display(dt.groupby('type_of_land')['price'].min())
print('='*30)
print(' Giá lớn nhất của mỗi loại nhà đất')
display(dt.groupby('type_of_land')['price'].max())
```

```python
print('Trung bình cộng số số phòng vệ sinh của mỗi phường')
display(dt.groupby('ward_name')['toilet'].mean())
print('='*30)
print('Trung bình cộng số số phòng ngủ của mỗi phường')
display(dt.groupby('ward_name')['bedroom'].mean())
print('='*30)
print('Trung bình cộng số số tầng của mỗi phường')
display(dt.groupby('ward_name')['floor'].mean())
```
