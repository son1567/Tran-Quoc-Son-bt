```python
n = 100 
a = 2 
b = 3 
c = 5 
count = 0
print("           2000đ   3000đ    5000đ")
for i in range(n // a + 1):
    for j in range(n // b + 1):
        for k in range(n // c + 1):
            if a*i + b*j + c*k == 100:
                count += 1
                print(" cách {0}    {1} tờ    {2} tờ    {3} tờ".format(count, i, j, k))
print("Tổng số cách:", count)               
```
