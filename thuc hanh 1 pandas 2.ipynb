```python
import pandas as pd
import numpy as np
print('Khai báo thành công')
```

```python
dt=pd.read_csv('FoodPrice_in_Turkey.csv')
dt.head()
```

```python
# Xóa các dòng có thuộc tính ProductID trùng nhau, giữ lại bản ghi cuối cùng, giữ chỉ số ban đầu của các dòng
dt=dt.drop_duplicates(['ProductId'], keep='last')
dt.head()
```

```python
# Xóa các dòng có thuộc tính ProductID trùng nhau, giữ lại bản ghi cuối cùng, thiết lập lại chỉ số
dt=dt.drop_duplicates(['ProductId'],keep='last').reset_index(drop=True)
dt.head()
```

```python
# Tách file chứa thông tin sản phẩm
dt_pro = dt.loc[:,['ProductId','UmId','UmName']]
dt_pro.head()
```

```python
# Tách file chứa thông tin giá
dt_pri = dt.loc[:,['ProductName','Price']]
dt_pri.head()
```

```python
# Tách file chứa thông tin giá với số dòng từ bản ghi 10 đến 20
dt_pri10 = dt.loc[10:20,['ProductId','Place','Month','Year','Price']]
dt_pri10
```

```python
#ghép các cột từ các file
dt=pd.concat([dt_pro,dt_pri],axis=1)
dt.head()
```
