```python
import pandas as pd
import numpy as np
print('Thành công')
```

```python
dt=pd.read_csv('OnlineRetail.csv',encoding='ISO-8859-1')
dt
```

```python
print('Số dòng là',dt.shape[0])
print('Số cột là',dt.shape[1])
```

```python
dt.info()
```

```python
dt['time']=dt['InvoiceDate'].str.split(' ').str[1]
dt['date']=dt['InvoiceDate'].str.split(' ').str[0]
dt.head()
```

```python
dt['Month']=pd.to_datetime(dt['date'],format='%m/%d/%Y').dt.month
dt.head()
```

```python
conditions = [(dt['Month'] >= 1) & (dt['Month'] <= 3),
              (dt['Month'] >= 4) & (dt['Month'] <= 6),
              (dt['Month'] >= 7) & (dt['Month'] <= 9),
              (dt['Month'] >= 10) & (dt['Month'] <= 12)] 
choices = [1,2,3,4]
dt.loc[:,'Previous'] = np.select(conditions, choices)
dt
```

```python
dt.loc[:,'Amount']=dt.loc[:,'Quantity']*dt.loc[:,'UnitPrice']
dt
```

```python
condition = [(dt['Month'] >= 10) & (dt['Month'] <= 12) & (dt['Country'] == 'United Kingdom'),
             (dt['Country'] == 'France')] 
choice = [10,5]
dt.loc[:,'Discount(%)'] = np.select(condition, choice , default=0)
dt
```

```python
dt.loc[:,'Total']=dt.loc[:,'Amount']-dt.loc[:,'Discount(%)']
dt
```
