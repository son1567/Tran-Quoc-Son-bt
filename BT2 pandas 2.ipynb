```python
al=dt.pivot_table(values='Quantity',index='Description',columns='Country',aggfunc='mean')
al
```

```python
c2=dt.pivot_table(values='Quantity',index='CustomerID',columns='StockCode',aggfunc=['min','max'])
c2
```
